```python
# Crear el gráfico de dispersión
plt.scatter(data['price'], data['city-mpg'])

# Establecer etiquetas y título
plt.xlabel('Precio')
plt.ylabel('Consumo (city-mpg)')
plt.title('Relación entre Precio y Consumo')

# Mostrar el gráfico
plt.show()
```

```python
# Calcular el precio promedio o mediano para cada categoría de body-style
body_style_prices = data.groupby('body-style')['price'].mean()

# Crear el gráfico de barras
plt.bar(body_style_prices.index, body_style_prices.values)

# Establecer etiquetas y título
plt.xlabel('Body Style')
plt.ylabel('Precio promedio')
plt.title('Relación entre Body Style y Precio')

# Mostrar el gráfico
plt.show()

```

Report on the relationship between body style, price and fuel consumption in our range of cars.

    Relationship between Price and Fuel Consumption:

In analyzing the relationship between price and fuel consumption of our automobiles, we have found that there is an inverse trend between these two variables. Lower-priced cars tend to have higher fuel consumption, while higher-priced cars tend to have lower fuel consumption. This relationship is consistent with the general characteristics of the automotive market, where more expensive vehicles tend to be equipped with more efficient engines and fuel-saving technologies.

    Body Style and Price:

In analyzing body style and its impact on the price of our automobiles, we have observed the following trends:

Convertibles and Hard Tops: These body styles stand out as having the highest average prices, exceeding $20,000. This can be attributed to the exclusive nature and specialized design of these vehicles, which often positions them as luxury or sporty options.

Sedans: Sedans rank second in terms of average price, with an approximate value of $15,000. Sedans are known for their versatility, comfort and interior space, making them a popular choice for many people.

Wagons: Wagons fall into a moderate average price range. They offer a balance between additional cargo space and family functionality, making them a practical choice for those seeking convenience and storage capacity.

Hatchbacks: Hatchbacks have the lowest average prices compared to other body styles. These compact, nimble vehicles are known for their fuel efficiency and ease of maneuverability, making them attractive to urban drivers and those looking for an economical vehicle.

These findings provide us with valuable information about market preferences and allow us to better understand the relationship between body style, price and fuel consumption. This information can be used to make strategic decisions regarding our automotive offerings and marketing strategies.
